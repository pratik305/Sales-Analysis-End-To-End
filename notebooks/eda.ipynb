## Star Schema Design

### FACT TABLES

### Fact_Sales (from orders + order_items)
Stores each product-level sales transaction.

- order_id: Unique identifier for each order
- order_item_id: Unique identifier for each item in an order
- product_id: Product sold
- website_session_id: Website session that led to the purchase
- created_at: Order creation date
- price_usd: Price of the product
- quantity: Quantity sold
- revenue: price_usd Ã— quantity

---

### Fact_Refunds (from order_item_refunds)
Stores refund events for sold products.

- order_item_id: Refunded order item
- refund_amount: Amount refunded
- refund_date: Date of refund

---

### DIMENSION TABLES

### Dim_Session (from website_sessions)
Stores marketing and device context of user visits.

- session_id: Unique website session identifier
- user_id: User identifier
- utm_source: Marketing source
- utm_campaign: Campaign name
- device_type: Device used
- created_at: Session start date

---

### Dim_Product (from products)
Stores product details.

- product_id: Unique product identifier
- product_name: Name of the product
- product_category: Category of the product
- created_at: Product launch date

---

### Dim_Date (derived table)
Stores calendar attributes for time-based analysis.

- date: Calendar date
- year: Year
- month: Month
- quarter: Quarter
- week: Week number
